```python
df[df['Price'] >= 15]['Author'].value_counts().sort_values(ascending=False).head(1)
```

```python
df.pivot_table(values='Reviews', index='Author', aggfunc='sum').sort_values('Reviews', ascending=False).head(5)
```

```python
df.pivot_table(values='Reviews', index='User Rating', aggfunc='sum').sort_index(ascending=False).head(5)
```

```python
# pd.cut(df['Price'], bins=4, duplicates='drop').value_counts()
price_bins = [0, 10, 20, 30, 40, 50, 110]
price_labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51+']
df['Price Range'] = pd.cut(df['Price'], bins=price_bins, labels=price_labels, right=False)
df.pivot_table(index='Price Range', columns='Genre', values='Name', aggfunc='count', fill_value=0)
```

```python
df[df['Genre'] == 'Fiction']['Year'].value_counts().sort_values(ascending=False).head(1)
```

```python
df[df['User Rating'] > 4.5]['Author'].value_counts().sort_values(ascending=False).head(1)
```

```python
plt.figure(figsize=(8, 6))
sns.boxenplot(x='Genre', y='User Rating', data=df)
plt.title('Distribution of User Ratings by')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Price', kde=True)
plt.title('Distribution of Book Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
