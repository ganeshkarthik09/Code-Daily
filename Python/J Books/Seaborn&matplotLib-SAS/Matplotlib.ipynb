```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

## Matplotlib Basics

```python
x = np.linspace(-5,5,22)
x
y = x ** 2
y  
```

```python
plt.plot(x,y)
plt.title("First Graph")
plt.xlabel("INT")
plt.ylabel("SQR")
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

### Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

### Object Oriented Way

```python

fig = plt.figure(figsize=(20,8),dpi = 60)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.1,0.2,.2,.2])
axis2.plot(x,y)
```

### Type Of Plots

```python
plt.scatter(x,y) # Scatter Plot
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data) # Histogram
```

```python
# Box Plot
plt.boxplot(x)
```

## Saving Plots

```python

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
