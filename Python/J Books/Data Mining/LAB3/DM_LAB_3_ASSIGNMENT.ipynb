```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris_sea.csv")
df
```

# **HISTOGRAM OF SEPAL_LENGTH**

```python
df2 = df.sepal_length
sns.histplot(df2, color = "skyblue", edgecolor = "black", kde = True)
plt.title("Histogram of sepal_length")
plt.ylabel("Frequency")
plt.grid()

plt.show()
```

# **COUNT OF SPECIES**

```python
sns.countplot(x = "species", data = df, hue = "species", palette = "viridis")
plt.title("Count of Species")

plt.show()
```

# **BOXPLOT OF SEPAL_LENGTH BY SPECIES**

```python
sns.boxplot(x = "species", y = "sepal_length", data = df, hue = "species", palette = "viridis")
plt.title("Boxplot of sepal_length by Species")

plt.show()
```

# **PAIRPLOT**

```python
sns.pairplot(data = df, hue = "species")
plt.title("Pair-plot")

plt.show()
```

# **CORRELATION MATRIX USING HEATMAP**

```python
df2 = df.drop(columns = "species")
corr_matrix = df2.corr()

sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation Matrix")

plt.show()
```

# **COMBINING ALL THE PLOTS**

```python
df2 = df.drop(columns="species")
corr_matrix = df2.corr()


fig, axes = plt.subplots(1, 4, figsize=(24, 5))

sns.histplot(df.sepal_length, color = "skyblue", edgecolor = "black", kde = True, ax = axes[0])
axes[0].set_title("Histogram of sepal_length")


sns.countplot(x = "species", data = df, hue = "species", palette = "viridis", ax = axes[1])
axes[1].set_title("Count of Species (Duplicate)")


sns.boxplot(x = "species", y = "sepal_length", data = df, hue = "species", palette = "viridis", ax = axes[2])
axes[2].set_title("Boxplot of Sepal Length")

sns.heatmap(corr_matrix, annot = True, ax = axes[3])
axes[3].set_title("Correlation Matrix")

sns.pairplot(data = df, hue = "species")
plt.title("Pairplot of Iris Dataset", y=1.02)

plt.tight_layout()
plt.show()
```
