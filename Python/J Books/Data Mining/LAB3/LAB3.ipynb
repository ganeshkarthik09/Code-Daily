```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(r'D:\Visual Studio Code\Practice\Python3\iris.data.csv', header=None, names=column_names)
```

```python
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr,annot = True,linecolor='Black',linewidths= 4,cmap='coolwarm')
```

```python
df.boxplot(
    column=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    patch_artist=True,
    boxprops=dict(facecolor='ORANGE'),
    medianprops = dict(color= 'red',linewidth = 4)
)
```

```python
column_names = ['sepal_length','sepal_width','petal_length','petal_width']
for column in column_names:
   plt.figure()
   df.boxplot(column=[column], vert=True, patch_artist=True, boxprops=dict(facecolor='orange', color='black'),
            medianprops=dict(color='red', linewidth=5),
            flierprops=dict(marker='o', color='orange', markersize=8))
   plt.title(f'Boxplot of {column}')
```

```python
df.hist(color = 'Blue',edgecolor='Black', grid =False,
	range=None, 
	density=False, 
	weights=None, 
	cumulative=False, 
	bottom=None, 
	histtype='bar', 
	align='mid',  
	log=False,  
	stacked=False, bins=10,
   rwidth=0.9,
   orientation='vertical',
   label='Normal Distribution' )
plt.suptitle('Histogram')
```

```python
sns.pairplot(data=df, hue='species')
plt.suptitle('Pair Plot of Iris Features', y=1)
plt.show()
```

```python
sns.histplot(df['sepal_width'],kde = True)
plt.title("Histogram of Sepal Length")
```

```python
sns.countplot(x = 'species',data = df,hue = 'species')
plt.title('Count of Species')
```
