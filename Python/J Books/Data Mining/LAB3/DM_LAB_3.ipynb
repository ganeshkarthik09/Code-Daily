```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df
```

# **HISTOGRAM**

```python
df.hist(figsize = (4,4), color = "red", edgecolor = "black", bins = 50)
plt.suptitle("HISTD")
plt.tight_layout()
plt.show()
```

# **BOXPLOT**

```python
df.boxplot(figsize = (5,5), color = "green")
plt.title("BOXPLOT")
plt.tight_layout()
plt.show()
```

# **PAIRPLOT**

```python
sns.pairplot(df, hue = "Species", palette = "husl")
```

# **CORRELATIONS**

```python
corr = df.select_dtypes(include = "number").corr()
sns.heatmap(corr, annot = True, linecolor = "red", linewidth = 4, cmap = "coolwarm")
```
