## Pandas Series
 A series is a one dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.

## Creating Series
There are multiple ways

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
my_array = np.array([10,10,30])
d = {1:10,2:50,3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(my_array)
```

```python
pd.Series(d)
```
