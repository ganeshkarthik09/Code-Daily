## Numpy Array and Basics

```python
import numpy as np
import time
```

### Creating array from list

```python
arr_1d = np.array([1,2,3,4])
print("1D arry :", arr_1d)
arr_2d = np.array([[1,2,3],[4,5,6]])
print("2D arry :", arr_2d)
```

### List VS Numpy Array

```python
py_list = [1,2,3]
print("Python list multiplication : ",py_list * 2)
np_array = np.array([1,2,3]) # element wise multiplication
print("Python array multiplication : ",np_array * 2)
start = time.time()
py_list= [i*2 for i in range(100000)]
print("\n List operation time : ",time.time()- start)
start = time.time()
np_array = np.arange(100000) * 2
print("\n Numpy  operation time : ",time.time()- start)
```

## Creating arrays from scratch

```python
zeros = np.zeros((3,4))
print(" Zero Array : ",zeros)
```

```python
ones = np.ones((2,2))
print("ones Array : ",ones)
```

```python
full = np.full((2,2),5)
print("full Array :\n ",full)
```

```python
rando = np.random.random((2,3))
print("rando Array :\n ",rando)
sequence = np.arange(0,11,2)
print("sequence Array :\n ",sequence)
```

## Vector , Matrix and Tensor

```python
vector = np.array([1,2,3])
print("Vector : ",vector)
matrix = np.array([[1,2,3],
                   [4,5,6]])
print("matrix : \n",matrix)
tensor = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print("tensor : \n",tensor)
```

### Array Properties

```python
arr = np.array([[1,2,3],[4,5,6.5]])
print("Shape : ", arr.shape)
print("Dimension : ",arr.ndim)
print("Size : " ,arr.size)
print("DType :" ,arr.dtype)
```

### Array reshaping

```python
arr = np.arange(12)
print("Original Array",arr)
reshaped = arr.reshape((3,4))
print("Reshaped Array : \n",reshaped)
flattened = reshaped.flatten()
print("flattened Array : \n",flattened)
# ravel(returns view, instead of copy)
raveled = reshaped.ravel()
print("raveled Array : \n",raveled)

#Transpose
transpose = reshaped.T
print("transpose Array : \n",transpose)
```
