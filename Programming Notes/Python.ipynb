## Loops and List Comprehensions

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
```

```python
#For loop in python includes `in`

for planet in planets:
    print(planet, end=' ')
```

*Tip: Here `end` is for ending outputs with spaces*

```python
s = 'steganograpHy is the practicE of conceaLing a file, message, image, or video within another fiLe, message, image, Or video.'
msg = ''
# print all the uppercase letters in s, one at a time
for char in s:
    if char.isupper():
        print(char, end='')    
```

`range()` a sequence of numbers.

```python
for i in range(5):
    print(i+5, end=' ')
```

```python
#While loops
i = 2
while i != 5:
    print(i)
    i += 1
```

Now, moving towards much important **List comprehensions**

```python
#Synatx ls = [f(x) for x in ls]
sq = [n**2 for n in range(5)]
sq
```

We can also add an `if` condition

```python
short_planets = [planet for planet in planets if len(planet) < 6]
short_planets
```

We can add a f(x) as well in above example.

```python
loud_short_planets = [planet.upper() + '!' for planet in planets if len(planet) < 6]
loud_short_planets
```

Below is an example with using function

```python
def count_negatives(nums):
    return sum([num < 0 for num in nums])

nums = [-2, -3, 2, 3, 4]
result = count_negatives(nums)
result
```

Now, we move on towards a very important data structure called Dictionaries.

```python
numbers = {'one':1, 'two':2, 'three':3}
```

Here, *one*, *two*, *three* are called keys and *1*, *2*,*3* are their values

Other name for dictonary is hash maps since we are mapping(very similar to functions in maths)

```python
numbers['one']
```

```python
#Lets add one more key-value pair
numbers['four'] = 4
numbers
```

Dictionaries also have dictionary comprehensions

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planets_dict = {i+1: planet for i, planet in enumerate(planets)}
planets_dict
```

*Tips: Enumerate is a function which returns index, value pair like below*

```python
for itr in enumerate(planets):
    print(itr)
```

```python
for k in numbers:
    print("{} = {}".format(k, numbers[k]))
```

Here it is iterating through keys and numbers[k] is returning value

Through `dict.keys()` and `dict.values()` we can get collection of keys and values 

```python
' '.join(sorted(planets_dict.values()))
```
